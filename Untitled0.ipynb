```python
import matplotlib.pyplot as plt   #Rename the package
import numpy as np
import random
from PIL import Image
```

Scatter plot

```python
X=[]
Y=[]

for i in range(100):
  x=i
  X.append(x)
  y=i+random.uniform(0,20) #random component
  Y.append(y)


figure,axis = plt.subplots(ncols=2, figsize=(16,8)) # 2 col, 1 row #fig size increase the size of the plot
axis[0].scatter(X,Y) #only 1 argument
axis[0].set_title("mosly linear with noise") #use matplotlib help

#If a want to see something here, like a figure I should uploaded to colab with the folder icon and then
with Image.open("/content/BOL_PE25_2SUV_50HR_2_03.tif") as image:
  image_data=np.array(image)

axis[1].imshow(image_data)
axis[1].set_title("SEM micrograph")

plt.show() #Take the last figure and added it the axis argument
```
